# Purpose
This notebook demonstrates 
- how to use the PTC data (e.g. which files to use) to get back the fits found in tsan et al. 2021: https://www.sciencedirect.com/science/article/pii/S0011227521000813.
- Checking the fit with the raw data.

The example shows here is for Stage 1 but the process can be apply to Stage 2. 

NOTE:
- *Stage1_data.pkl* and *Stage2_data.pkl* contain the tilt data and the high density load curves at 0, 27.5, 36, 41, and 48 degrees for their respective temperature stages. 
- *Stage1_THLC_fitresults.txt* and *Stage2_THLC_fitresults.txt* are the fit results.

```python
import numpy as np
import matplotlib.pyplot as plt
from lmfit import Parameters, minimize
from matplotlib.pyplot import cm 
import pickle 
```

```python
########### Functions to extract the fit parameters from load curves ########### 
####### 1. Create an objective function to calculate the residues ###########
"""
The idea is to have some inital guesses for a, b, and c for all 6 sets of data.
Then plot 6 models of the exponential function based on these initial values. 
Find the difference between the real data and the model. 
For the residues, don't need to make a new x, can just use the corresponding x of each data set.
So need to combine all 5 data sets into 1. 
"""
def exponential_func(x, a, b, c):
    return a*np.exp(b*x)+c

def expo_set(params, i, x):
    a = params['a%i' % (i+1)]
    b = params['b%i' % (i+1)]
    c = params['c%i' % (i+1)]
    return exponential_func(x, a, b, c)

def objective(params, x, data):
    for i, d in enumerate(data):
        diff = d - expo_set(params, i, x[i])
        if i == 0:
            resid = diff
        else:
            resid = np.concatenate((resid, diff))
    arr = np.array((resid))
    return arr


def makeparams1(sz, vals, fparam=None):
    """
    vals (1D array) = the initial guess values for all the parameters
    """
    fit_params = Parameters()
    for i in range(1, sz+1):
        fit_params.add('a%i' % i, value = vals[0], min = 0)
        fit_params.add('b%i' % i, value = vals[1], min = 0)
        fit_params.add('c%i' % i, value = vals[2], min=0)
    if fparam != None:
        for fp in fparam:
            skip = int(fp[-1])
            for i in range(1, sz+1):
                if i == skip: continue
                fit_params[fp[:-1]+'%i' % i].expr = fp
    return fit_params

######### 2. Fixed b: put all into 1 big dictionary #########
def Fixedb_Fitall(data_in, n, guess, fp=None, printresult=True):
    """
    n = number of data sets 
    guess = array of initial guesses for the fit parameters
    fp = string of parameter to fixed across data sets
    """
    x = []
    data = []
    pows1 = list(data_in.keys())
    for i, p1 in enumerate(pows1):
        pows2 = list(data_in[p1].keys())
        for j, p2 in enumerate(pows2):
            x.append(data_in[p1][p2][0]) # angle
            data.append(data_in[p1][p2][1]) #T data
            if i == 0 and j ==0:
                std = data_in[p1][p2][3] #T std
            else:
                std = np.concatenate((std, data_in[p1][p2][3]))
    fparams = makeparams1(n, guess, fp)
    results = minimize(objective, fparams, args=(x, data))
    if printresult == True:
        results.params.pretty_print()
    
    # get best-fits
    cov = results.covar
    errs = np.sqrt(np.diag(cov))
    best_vals = results.params.valuesdict()
    a_vals = []
    c_vals = []
    b_vals = []
    for i in range(1, n+1):
        if i == 1: 
            b_vals.append([best_vals['b%i' %(i)], errs[1]])
            a_vals.append([best_vals['a%i' %(i)], errs[0]])
            c_vals.append([best_vals['c%i' %(i)], errs[2]])
        
        else:
            a_vals.append([best_vals['a%i' %(i)], errs[i*2-1]])
            c_vals.append([best_vals['c%i' %(i)], errs[i*2]])
            b_vals.append([best_vals['b%i' %(i)], errs[1]])
       
    a_vals = np.array((a_vals))
    b_vals = np.array((b_vals))
    c_vals = np.array((c_vals))
    
    # chi square and reduced chi sq
    res = results.residual
    chisq = np.sum(res**2 / std**2)
    redchi = chisq/results.nfree
    return a_vals, b_vals, c_vals, chisq, redchi
```

```python
########### Fitting Functions  ########### 
def fitline(x, fn, params, npoints = 100, overwrite_max= False, new_max = 0):
    if np.size(x) == 1:
        xfit = x
    else:
        ind_min = np.argmin(x)
        ind_max = np.argmax(x)
        if overwrite_max: x_max = new_max
        else: x_max = x[ind_max]
        xfit = np.linspace(x[ind_min], x_max, npoints)
    yfit = fn(xfit, params)
    return np.array([xfit, yfit]).T
def expo_fn(x, params):
    return params[0]*np.exp(params[1]*x)+params[2]

## finding the correct fit parameters that correspond to the power indicated
def extract_params(fit_results, p1, p2):
    p1_arr = fit_results[:, 0]
    p2_arr = fit_results[:, 1]
    inds = np.where((p1_arr == p1) & (p2_arr == p2))
    params = [fit_results[:, 2][inds][0], fit_results[:, 3][inds][0], fit_results[:, 4][inds][0]] #a, b, c
    return params
    
```

## Stage 1

```python
### load data 
with open('Stage1_data.pkl', 'rb') as f:
    T1data = pickle.load(f)
# load fit results
fit_results1 = np.loadtxt('Stage1_THLC_fitresults.txt')
```

```python
### checking out what is in Stage1_data.pkl file. 
"""
This is a dictionary with the keys in P1 and P2, respectively.
"""
T1data
```

```python
P1s = np.array(list(T1data.keys()))
P2s = np.array(list(T1data[0].keys()))
print(P1s, P2s)
```

```python
### fit all data: tilt + HDLC and fixed b
"""
The fit results here are what saved in Stage1_THLC_fitresults.txt.
"""
init_vals = np.array((0.0183, 0.119, 29.49))
n = 48
fp = ['b1']
a_vals, b_vals, c_vals, chi, redchi = Fixedb_Fitall(T1data, n, init_vals, fp)
print(chi, redchi)
```

```python
################################ Plotting #####################################
#### plot all the 40Ks ####

plt.figure(figsize=(20,30)) 
for j, p1 in enumerate(P1s):
    plt.subplot(4,2,j+1)
    col = iter(cm.rainbow(np.linspace(0,1,len(P2s))))
    for i, p in enumerate(P2s):
        co = next(col)
        m2 = fitline(T1data[p1][p][0], expo_fn, extract_params(fit_results1, p1, p))
        plt.plot(m2[:, 0], m2[:, 1], '-', color = co, linewidth = 2)
        plt.plot(T1data[p1][p][0], T1data[p1][p][1], 'o', color = co, label='$P_2$ = %s W' %str(p))
        plt.errorbar(T1data[p1][p][0], T1data[p1][p][1], T1data[p1][p][3], T1data[p1][p][2], fmt='o', color = co)

    plt.xlabel(r'Angle ($^{\circ}$)', fontsize=18)
    plt.ylabel('Stage-1 Temperature (K)', fontsize=18)
    plt.title('$P_1$ = %s W'% str(p1), fontsize=20)
    plt.tick_params(axis='both', labelsize = 18)
    plt.legend(fontsize = 15)
    plt.grid()
    plt.xlim(-0.2, 57)
    plt.ylim(24, 58)
plt.subplots_adjust(hspace=.25)
```
